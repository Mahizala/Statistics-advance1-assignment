**Q4.> What are probability distribution functions (PDF)?**

-->In probability theory, a probability density function (PDF), density function, or density of an absolutely continuous random variable, is a function whose value at any given sample (or point) in the sample space (the set of possible values taken by the random variable) can be interpreted as providing a relative likelihood that the value of the random variable would be equal to that sample.[2][3] Probability density is the probability per unit length, in other words, while the absolute likelihood for a continuous random variable to take on any particular value is 0 (since there is an infinite set of possible values to begin with), the value of the PDF at two different samples can be used to infer, in any particular draw of the random variable, how much more likely it is that the random variable would be close to one sample compared to the other sample.

**Q5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

-->The fundamental distinction between a Probability Density Function (PDF) and a Cumulative Distribution Function (CDF) lies in what each function represents regarding the probability of a random variable. The PDF, primarily used for continuous random variables, describes the likelihood of a random variable taking on a specific value within a given range. Essentially, it indicates the "density" of probability at each point. In contrast, the CDF, applicable to both discrete and continuous variables, provides the probability that a random variable will take on a value less than or equal to a specified value. Therefore, the CDF is a cumulative measure, showing the accumulated probability up to a certain point, while the PDF depicts the probability density at individual points or within intervals. Another way to think of it is that the CDF is the integral of the PDF.

**Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cdf_discrete_uniform(low, high, size):
    """
    Calculates and plots the cumulative distribution function (CDF) of a discrete uniform distribution.

    Args:
      low: The lower bound of the uniform distribution.
      high: The upper bound of the uniform distribution.
      size: The number of random samples to generate.
    """
    data = np.random.randint(low, high + 1, size)
    values, counts = np.unique(data, return_counts=True)
    probabilities = counts / size
    cdf = np.cumsum(probabilities)

    plt.figure(figsize=(8, 6))
    plt.step(values, cdf, where='post')
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
    plt.grid(True)
    plt.show()

#Example usage
plot_cdf_discrete_uniform(1, 6, 1000)
```

**Q9.Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def z_scores(data):
    return (data - np.mean(data)) / np.std(data, ddof=1)

def plot_z_scores(z):
    plt.hist(z, bins=20, density=True, alpha=0.6, color='b', label='Z-scores')
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, stats.norm.pdf(x), 'r-', label='Standard Normal Curve')
    plt.xlabel('Z-score'), plt.ylabel('Density'), plt.title('Histogram of Z-scores')
    plt.legend(), plt.grid(), plt.show()

data = np.random.normal(50, 10, 100)
plot_z_scores(z_scores(data))
```

**Q18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.**

```python

import scipy.stats as stats

def calculate_z_score(data_point, population_mean, population_std_dev):
    """Calculates the Z-score for a given data point.

    Args:
        data_point: The data point for which to calculate the Z-score.
        population_mean: The mean of the population.
        population_std_dev: The standard deviation of the population.

    Returns:
        The Z-score.
    """
    z_score = (data_point - population_mean) / population_std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """Compares a Z-score to the standard normal distribution.

    Args:
        z_score: The Z-score to compare.
    """
    p_value = stats.norm.cdf(z_score)  # Calculate cumulative probability
    print(f"Z-score: {z_score}")
    print(f"Cumulative Probability (P-value): {p_value}")

    # Interpretation
    if abs(z_score) > 1.96:
      print("The data point is significantly different from the population mean (at the 5% level).")
    else:
      print("The data point is not significantly different from the population mean (at the 5% level).")


# Example Usage
data_point = 75
population_mean = 70
population_std_dev = 5

z = calculate_z_score(data_point, population_mean, population_std_dev)
compare_to_standard_normal(z)
```

**Q19.Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def hypothesis_test_z(sample, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
    reject_null = p_value < alpha

    return {
        "z_statistic": z_statistic,
        "p_value": p_value,
        "reject_null": reject_null
    }

# Example usage
sample_data = np.array([65, 70, 72, 68, 75, 71, 69, 73, 78, 76])
population_mean_hypothesized = 70
population_std_dev = 5
results = hypothesis_test_z(sample_data, population_mean_hypothesized, population_std_dev)
results
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(dataset1, dataset2):

    std_dev1 = np.std(dataset1)
    std_dev2 = np.std(dataset2)

    print(f"Standard Deviation of Dataset 1: {std_dev1}")
    print(f"Standard Deviation of Dataset 2: {std_dev2}")

    plt.figure(figsize=(6, 4))
    plt.boxplot([dataset1, dataset2], labels=['Dataset 1', 'Dataset 2'])
    plt.ylabel('Value')
    plt.title('Comparison of Standard Deviations')
    plt.show()

    # Example usage
dataset1 = np.random.normal(loc=50, scale=10, size=100)
dataset2 = np.random.normal(loc=60, scale=5, size=100)
compare_standard_deviations(dataset1, dataset2)
```

**Q26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.**

```python
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate the range
data_range = np.ptp(data)
# Calculate the interquartile range (IQR)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {iqr}")
```

**Q27.Implement Z-score normalization on a dataset and visualize its transformation.**

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalize(data):
    data = np.array(data)  # Convert to NumPy array if necessary
    normalized_data = (data - np.mean(data)) / np.std(data)
    return normalized_data

# Example usage:
data = np.random.normal(loc=50, scale=10, size=100)

# Apply Z-score normalization
normalized_data = z_score_normalize(data)

plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.hist(data, bins=20, alpha=0.6, color='purple',edgecolor='black', label='Original Data')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()


plt.subplot(2, 1, 2)
plt.hist(normalized_data, bins=20, alpha=0.6, color='orange',edgecolor='black',label='Normalized Data')
plt.title('Normalized Data Distribution (Z-score)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

**Q28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.**

```python
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    return {"skewness": skewness, "kurtosis": kurtosis}


# Example usage:
data = np.random.normal(loc=0, scale=1, size=1000)
results = calculate_skewness_kurtosis(data)
results
```
